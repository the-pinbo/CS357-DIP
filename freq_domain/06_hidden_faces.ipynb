```python
from skimage import filters
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_p = cv2.imread('./image/pranav.jpeg',0)
small_to_large_image_size_ratio = 0.125
img_p = cv2.resize(img_p, # original image
                       (0,0), # set fx and fy, not the final size
                       fx=small_to_large_image_size_ratio, 
                       fy=small_to_large_image_size_ratio, 
                       interpolation=cv2.INTER_NEAREST)
plt.imshow(img_p,cmap='grey')
plt.show()
```

```python
img_u = cv2.imread('./image/utkarsh.jpeg',0)
small_to_large_image_size_ratio = 0.125
img_u = cv2.resize(img_u, # original image
                       (0,0), # set fx and fy, not the final size
                       fx=small_to_large_image_size_ratio, 
                       fy=small_to_large_image_size_ratio, 
                       interpolation=cv2.INTER_NEAREST)
plt.imshow(img_u,cmap='grey')
plt.show()
```

```python
ut_filter = filters.butterworth(img_u,cutoff_frequency_ratio=.05, order=2, high_pass=False)
plt.imshow(ut_filter,cmap='grey')
plt.axis(False)
plt.show()

p_filter = filters.butterworth(img_p,cutoff_frequency_ratio=.05, order=2, high_pass=True)
plt.imshow(p_filter,cmap='grey')
plt.axis(False)
plt.show()

hidden_img = np.zeros(img_p.shape)
cv2.accumulate(ut_filter,hidden_img)
cv2.accumulate(p_filter, hidden_img)
plt.imshow(hidden_img,cmap='grey')
plt.axis(False)
plt.show()
```
